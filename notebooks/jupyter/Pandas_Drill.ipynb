```python
# Step 0. Import libraries and custom modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.animation as animation
import seaborn as sns
plt.rcParams['animation.html'] = 'jshtml'
from matplotlib.dates import DateFormatter
```

```python
# Step 1. Load data, show info and sample
# 1.1 Read data
df_raw = pd.read_excel(
   'https://github.com/datagy/mediumdata/raw/master/sample_pivot.xlsx', 
   parse_dates=['Date']
)
df_raw.info()
```

```python
# 1.1 Show sample
df_raw.sample(5, random_state=1234)
```

```python
# Step 2. Transform to standard format
df = (
    df_raw
    .copy()
    .set_axis(
        df_raw
        .columns.str.lower()
        .str.replace(' ','_')
        .str.replace(r'\W','',regex=True)
        .str.slice(0,40), axis=1
    )
    .astype({'region':'category', 
             'type':'category'})
    .assign(
        date = lambda x: pd.to_datetime(x['date'], format='%d/%m/%Y')
    )
    #.iloc[:, [-1] + [x for x in range(df_raw.shape[1]) if x!=4]]
    .set_index('date')
)
```

```python
# Step 3. Perform busines analysis
# 3.1 Plot sales per day
df['sales'].resample('D').sum().plot(title='Sales per day', c='purple')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# Plot sales per month
(
    df['sales'].resample('ME').sum()
    .plot(kind='bar', title='Sales per month', color='purple')
)
(
    plt.gca().xaxis
    .set_ticklabels(df['sales'].resample('ME').sum().index.to_period('M'))
)
plt.grid(axis='y')
plt.show()
```

```python
monthly = (
    df.groupby('region', observed=False)['sales']
    .resample('ME').sum().reset_index()
)
plt.figure(figsize=(6,4))
sns.lineplot(data=monthly, x='date', y='sales', 
             hue='region', palette='Set1', marker='X')
plt.grid(True)
plt.legend(loc='upper center', ncol=4, bbox_to_anchor=(0.5, 1.12))
plt.show()
```
